# Expense Tracker

```python
expenses = []

print("===== EXPENSE TRACKER =====")
print("Expense tracker started successfully!")
```

# Add Expense

```python
date = input("Enter the date (YYYY-MM-DD): ")
category = input("Enter the category: ")
description = input("Enter the description: ")
amount = float(input("Enter the amount: "))

expense = {
    "date": date,
    "category": category,
    "description": description,
    "amount": amount
}

expenses.append(expense)

print("\nExpense added successfully! ✅")
print(expense)
```

# View All Expenses

```python
print("\n===== ALL EXPENSES =====")

for i, expense in enumerate(expenses, start=1):
    print(f"\nExpense {i}")
    print(f"Date: {expense['date']}")
    print(f"Category: {expense['category']}")
    print(f"Description: {expense['description']}")
    print(f"Amount: ₹{expense['amount']}")
```

# Add Multiple Expenses

```python
while True:

    print("\n===== ADD EXPENSE =====")

    date = input("Enter the date (YYYY-MM-DD): ")
    category = input("Enter the category: ")
    description = input("Enter the description: ")

    while True:
        try:
            amount = float(input("Enter the amount: "))

            if amount > 0:
                break
            else:
                print("Amount must be greater than 0.")

        except ValueError:
            print("Please enter a valid amount.")

    expense = {
        "date": date,
        "category": category,
        "description": description,
        "amount": amount
    }

    expenses.append(expense)

    print("\nExpense added successfully! ✅")

    choice = input("\nDo you want to add another expense? (yes/no): ")

    if choice.lower() != "yes":
        break
```

# Calculate Total Expenses

```python


total_expenses = 0

for expense in expenses:
    total_expenses += expense["amount"]

print("\n===== TOTAL EXPENSES =====")
print(f"Total Amount Spent: ₹{total_expenses:.2f}")
```

# Category-wise Expense Analysis


```python
category_totals = {}

for expense in expenses:
    category = expense["category"]
    amount = expense["amount"]

    if category in category_totals:
        category_totals[category] += amount
    else:
        category_totals[category] = amount

print("\n===== CATEGORY-WISE EXPENSES =====")

for category, total in category_totals.items():
    print(f"{category}: ₹{total:.2f}")
```

# Category-wise Expense Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))

plt.bar(
    category_totals.keys(),
    category_totals.values()
)

plt.title("Expense by Category")
plt.xlabel("Category")
plt.ylabel("Amount Spent (₹)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Save Expenses to CSV

```python
import pandas as pd

expenses_df = pd.DataFrame(expenses)

expenses_df.to_csv("expenses.csv", index=False)

print("Expenses saved successfully to expenses.csv ✅")

print("\nSaved Expense Data:")
print(expenses_df)
```

# Search Expenses by Category

```python
search_category = input("Enter the category to search: ")

found_expenses = []

for expense in expenses:
    if expense["category"].lower() == search_category.lower():
        found_expenses.append(expense)

print("\n===== SEARCH RESULTS =====")

if found_expenses:
    for expense in found_expenses:
        print(f"Date: {expense['date']}")
        print(f"Category: {expense['category']}")
        print(f"Description: {expense['description']}")
        print(f"Amount: ₹{expense['amount']}")
        print("-" * 30)
else:
    print("No expenses found for this category.")
```

```python
def category_pie_chart():
    category_totals = {}

    for expense in expenses:
        category = expense["category"]
        amount = expense["amount"]

        category_totals[category] = category_totals.get(category, 0) + amount

    if not category_totals:
        print("\nNo expenses available for visualization.")
        return

    plt.figure(figsize=(8,8))

    plt.pie(
        category_totals.values(),
        labels=category_totals.keys(),
        autopct="%1.1f%%",
        startangle=90
    )

    plt.title("Expense Distribution by Category")
    plt.show()
```
