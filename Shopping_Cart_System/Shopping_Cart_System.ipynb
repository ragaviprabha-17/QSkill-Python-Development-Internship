# Product Catalogue

```python
products = {
    "Laptop": 55000,
    "Smartphone": 25000,
    "Tablet": 18000,
    "Smartwatch": 5000,
    "Headphones": 2000,
    "Bluetooth Speaker": 3000,
    "Keyboard": 1500,
    "Mouse": 800,
    "Monitor": 12000,
    "Webcam": 2500,
    "Power Bank": 1800,
    "USB Cable": 500,
    "Laptop Stand": 1200,
    "Printer": 15000,
    "External Hard Drive": 6000
}

print("===== PRODUCT CATALOGUE =====")

for product, price in products.items():
    print(f"{product}: ₹{price}")
```

# Shopping Cart

```python
cart = {}

print("\n===== ADD PRODUCTS TO CART =====")

product_name = input("Enter the product name: ")

if product_name in products:
    quantity = int(input("Enter the quantity: "))
    cart[product_name] = quantity
    print(f"{quantity} x {product_name} added to your cart.")
else:
    print("Product not available.")
```

```python
cart = {}

print("\n===== ADD PRODUCTS TO CART =====")

while True:
    product_name = input(
        "\nEnter the product name (or type 'done' to finish): "
    ).strip()

    if product_name.lower() == "done":
        break

    if product_name in products:

        while True:
            try:
                quantity = int(input("Enter the quantity: "))

                if quantity > 0:
                    break
                else:
                    print("Quantity must be greater than 0.")

            except ValueError:
                print("Please enter a valid number for the quantity.")

        if product_name in cart:
            cart[product_name] += quantity
        else:
            cart[product_name] = quantity

        print(f"{quantity} x {product_name} added to your cart.")

    else:
        print("Product not available.")

print("\n===== YOUR SHOPPING CART =====")

for product, quantity in cart.items():
    print(f"{product}: {quantity}")
```

# Calculate Total Price

```python
total_price = 0

print("\n===== BILL SUMMARY =====")

for product, quantity in cart.items():
    price = products[product]
    item_total = price * quantity

    print(f"{product} x {quantity} = ₹{item_total}")

    total_price += item_total

print("\nTotal Price: ₹", total_price)
```

# Discount System


```python
if total_price >= 50000:
    discount_percentage = 20
elif total_price >= 25000:
    discount_percentage = 10
elif total_price >= 10000:
    discount_percentage = 5
else:
    discount_percentage = 0

discount_amount = total_price * discount_percentage / 100
final_price = total_price - discount_amount

print("\n===== DISCOUNT DETAILS =====")
print("Original Price: ₹", total_price)
print("Discount:", discount_percentage, "%")
print("Discount Amount: ₹", discount_amount)
print("Final Price: ₹", final_price)
```

# Final Checkout Receipt


```python
print("\n" + "=" * 40)
print("          FINAL SHOPPING RECEIPT")
print("=" * 40)

print("\nItems Purchased:")

for product, quantity in cart.items():
    price = products[product]
    item_total = price * quantity

    print(f"{product}")
    print(f"  Quantity: {quantity}")
    print(f"  Price: ₹{price}")
    print(f"  Total: ₹{item_total}")

print("\n" + "-" * 40)
print(f"Original Price: ₹{total_price:.2f}")
print(f"Discount: {discount_percentage}%")
print(f"Discount Amount: ₹{discount_amount:.2f}")
print(f"Final Price: ₹{final_price:.2f}")
print("-" * 40)

print("Thank you for shopping with us! 🛒")
```
