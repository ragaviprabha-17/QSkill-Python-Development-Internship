```python
average_age = df["age"].mean()

print("Average Age of Passengers:", round(average_age, 2))
```

```python
plt.figure(figsize=(7,5))

df["class"].value_counts().plot(kind="bar", color=["skyblue","orange","green"])

plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.scatter(df["age"], df["fare"], alpha=0.6)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare Paid")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="survived", data=df)

plt.title("Passenger Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="sex", data=df)

plt.title("Male vs Female Passengers")

plt.show()
```

**Observations**
The Titanic dataset contains 891 passenger records.
The average passenger age is approximately 29.7 years.
Most passengers travelled in Third Class.
There are missing values in the Age, Deck, Embarked, and Embark Town columns.
There is no strong linear relationship between passenger age and ticket fare.
The heatmap shows that numerical features have only weak to moderate correlations.
The dataset is suitable for learning data analysis and visualization using Python.
